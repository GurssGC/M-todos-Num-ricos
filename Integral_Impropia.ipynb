```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad
```

```python
# Función a integrar
def f(x):
    return np.exp(-x)/np.sqrt(x)
```

```python
a = 0 # Límite inferior
b = np.Infinity # Límite superior

x = np.linspace(0.1, 4, 100)
y = f(x)
plt.plot(x,y)
plt.show()
```

```python
# Llamada a la funcion quad de scipy
I1 = quad(f, a, b)
print('quad = ', I1)
```

```python
# Función a integrar
def g(x):
    return 1/(1+x**2)

a = -np.Infinity # Límite inferior
b = np.Infinity # Límite superior
    
# Llamada a la funcion quad de scipy

I2 = quad(g, a, b)
print('quad = ', I2)
    
# Gráfica de g(x)
x = np.linspace(-5.0, 5.0, 100)
y = g(x)
plt.plot(x,y)
plt.show()
```

```python
# Función a integrar
def h(x):
    return 1/np.sqrt(x)

a = 0.0 # Límite inferior
b = 1.0 # Límite superior
    
# Llamada a la funcion quad de scipy
I3 = quad(h, a, b)
print('quad = ', I3)
    
# Gráfica de h(x)
x = np.linspace(a, 1.0, 100)
y = h(x)
plt.plot(x,y)
plt.show()
```

```python
# Función a integrar

def j(x):
    return 1/np.sqrt(np.pi)*np.exp(-x**2)

a = -np.Infinity # Límite inferior
b = np.Infinity # Límite superior
    
# Llamada a la funcion quad de scipy
I4 = quad(j, a, b)
print('quad = ', I4)
    
# Gráfica de h(x)
x = np.linspace(-5.0, 5.0, 100)
y = j(x)
plt.plot(x,y)
plt.show()
```
