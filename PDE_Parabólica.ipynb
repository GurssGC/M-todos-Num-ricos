```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Par√°metros de la barra.
L = 10                # Longitud de la barra
dx = 1                # delta x
dt = 0.1              # delta t
alfa = 0.835          # Coeficiente de difusividad termica
lamda = alfa*dt/dx**2 # Valor de lamda
Ti = 100              # Frontera izquierda
Td = 50               # Frontera derecha
T0 = 10              # Valores intermedios
```

```python
n = 200               # Vector tiempo
m = L//dx             # Vector dimension
u = np.zeros((m+1,n)) # Matriz de resultados
u[0,] = Ti            # Valor en la frontera izquierda
u[m,] = Td            # Valor en la frontera derecha
u[1:m,] = T0          # Valores intermedios 
```

```python
for ti in range(n-1):
        u[1:m,ti+1]=[u[xi,ti]+lamda*(u[xi-1,ti]-2*u[xi,ti]+u[xi+1,ti])\
          for xi in np.arange(1,m)]
```

```python
#np.set_printoptions(precision=2)
print (u)
```

```python
x = np.linspace(0,L,m+1)
plt.figure(figsize = (8, 6))
plt.plot(x ,u[:,::10])
plt.show()
```
