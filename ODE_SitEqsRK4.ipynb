```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Metodo de Runge-Kutta
def RK4(f, g, x0, y0, z0, x1, n):
    h = (x1-x0)/n      #tamano de paso
    xi = np.zeros(n+1) #vector de x variable independiente
    yi = np.zeros(n+1) #vector de y variable dependiente
    zi = np.zeros(n+1)
    xi[0] = x0         #valor inicial var. independiente
    yi[0] = y0         #condicion inicial var. dependiente
    zi[0] = z0
    
    for i in range(n):
        k1y = f(xi[i], yi[i], zi[i])
        k1z = g(xi[i], yi[i], zi[i])
        
        k2y = f(xi[i]+h/2, yi[i]+k1y*h/2, zi[i]+k1z*h/2)
        k2z = g(xi[i]+h/2, yi[i]+k1y*h/2, zi[i]+k1z*h/2)
        
        k3y = f(xi[i]+h/2, yi[i]+k2y*h/2, zi[i]+k2z*h/2)
        k3z = g(xi[i]+h/2, yi[i]+k2y*h/2, zi[i]+k2z*h/2)
        
        k4y = f(xi[i]+h, yi[i]+k3y*h, zi[i]+k3z*h)
        k4z = g(xi[i]+h, yi[i]+k3y*h, zi[i]+k3z*h)
        
        yi[i+1] = yi[i] + h*(k1y/6 + k2y/3 + k3y/3 + k4y/6)
        zi[i+1] = zi[i] + h*(k1z/6 + k2z/3 + k3z/3 + k4z/6)         
        xi[i+1] = xi[i] + h
        
    return xi, yi, zi    #Vector de valores calculados
```

```python
def f(x,y0,y1):
    F = 3*y0+8*y1
    return F

def g(x,y0,y1):
    F = -y0-3*y1
    return F
```

```python
x0 = 0
x1 = 10
y0 = 6
z0 = -2
n = 30
```

```python
X, Y, Z = RK4(f, g, x0, y0, z0, x1, n)
```

```python
plt.figure( figsize = (8, 6) )
plt.plot(X, Y, label='funcion 1')
plt.plot(X, Z, label='funcion 2')
plt.legend()
plt.show()
```

```python
def f(x, y0, y1):
    F = -4.0*(y0 + y1)
    return F

def g(x, y0, y1):
    F = y0
    return F
```

```python
x0 = 0
y0 = 1
z0 = 0
x1 = 10
n = 1000
```

```python
X, Y, Z = RK4(f, g, x0, y0, z0, x1, n)
```

```python
def fa(t):
    F = (1-2*t)*np.exp(-2*t)
    return F

def ga(t):
    F = t*np.exp(-2*t)
    return F
```

```python
t = np.linspace(0, 10, 1000)
ya = fa(t)
za = ga(t)
```

```python
plt.figure( figsize = (8, 6) )
plt.plot(X, Y, '.--')
plt.plot(t, ya)
plt.show()
```
