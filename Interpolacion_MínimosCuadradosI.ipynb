# ## Mínimos cuadrados, regresión con polinomio grado 1. I 

```python
# Librerías
import numpy as np
import sympy as sym
import matplotlib . pyplot as plt
from numpy.polynomial.polynomial import Polynomial
from scipy.interpolate import lagrange
from scipy import stats
```

```python
# Datos experimentales
ti = np.array ([1 ,2 ,3 ,4 ,5 ,6 ,7])
xi = np.array ([0.5 ,2.5 ,2.0 ,4.0 ,3.5 ,6.0 ,5.5])
```

$\textbf{Interpolación Polinomial por Lagrange}$

```python
# Interpolación de Lagrange de grado 6
poly = lagrange(ti, xi)
Polynomial(poly.coef[::-1]).coef
```

### Mínimos Cuadrados para polinomio de grado 1

```python
# Regresión lineal por mínimos cuadrados
a1, a0, r , p ,std_err = stats.linregress(ti, xi)
```

```python
# Datos de salida
print( 'Ordenada al origen a0 = ', a0 )
print( 'Pendiente a1 = ', a1 )
print( 'Coeficiente de correlacion r = ', r )
print( 'Coeficiente de determinación r^2 = ', r**2 )
```

```python
# Polinomio grado 1
t = sym.Symbol('t')
xs = a0 + a1*t
print( 'x(t)= ',xs )
```

```python
# Polinomio grado 1
def f(t):
    return a0 + a1*t

t = np.linspace(1, 7, 100)
x = f(t)
```

```python
t = np.linspace(1, 7, 100)
# Grafica
plt.figure(figsize=(8,6))
plt.scatter( ti, xi, color='b', label="Datos" )
plt.plot(t, Polynomial(poly.coef[::-1])(t), '--',color='r', label='Polinomio')
plt.plot(t, x, color='g', label='Regresión lineal')
plt.xlabel( 'Tiempo ($t$)')
plt.ylabel( 'Posición $y(t)$')
plt.title( 'Movimiento Uniforme')
plt.legend()
plt.show()
```
