```python
# Librerías 
import numpy as np
```

```python
# Regla de Integración
def RTrapecio(f, a, b, n):
    h  = (b - a) / n
    xs = np.linspace(a, b, n+1)
    ys = f(xs)
    r  = np.absolute( h*( ys[0] + 2*sum(ys[1:n]) + ys[n] ) / 2 )
    return r
```

```python
# Función a Integrar
def f(x):
    return 3*x**2+1
```

```python
a = 0 # Límite Inferior
b = 2 # Límite Superior
n = 1000 # Número de Trapecios
area = RTrapecio(f, a, b, n)
print('Integral = ',area)
```

```python
import matplotlib.pyplot as plt
x = np.linspace(0, 2, 100)
y = f(x)
plt.figure(figsize=(8, 8))
plt.plot(x, y)
plt.plot(x, x*0+1)
plt.grid()
plt.show()
```
