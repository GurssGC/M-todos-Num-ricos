```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Metodo de Euler Modificado
def EulerMod(f,x0,y0,x1,n):
    h=(x1-x0)/n      #tamano de paso
    xi=np.zeros(n+1) #vector de x variable independiente
    yi=np.zeros(n+1) #vector de y variable dependiente
    xi[0]=x0         #tiempo iniicial
    yi[0]=y0         #concentracion inicial
    
    for i in range(n):
        xi[i+1]=xi[i]+h
        yi[i+1]=yi[i]+h*f(xi[i],yi[i])
        yi[i+1]=yi[i]+h*(f(xi[i],yi[i])+f(xi[i+1],yi[i+1]))/2
        
    return xi,yi    #Vector de valores calculados
```

##### Dada la ecuación 
\begin{equation}
\left(1+e^{x}\right)y\frac{dy}{dx}=e^{x},
\end{equation}
##### hallar la solución que satisface la condición inicial 
\begin{equation}
\left. y\right|_{x=0}=1.
\end{equation}
##### Solución analítica
\begin{equation}
y=\sqrt{1+\ln{\left(\frac{1+e^x}{2}\right)^2}}.
\end{equation}

```python
def f(x,y):
    return np.exp(x)/(y*(1+np.exp(x)))

def f2(x):
    return np.sqrt(1+np.log(((1+np.exp(x))/2)**2)) ## Solución analítica
```

```python
x0 = 0.0   #valor inicial de tiempo
y0 = 1.0   #valor inicial de la concentracion
x1 = 200.0  #valor final del tiempo
n = 50     #numero de pasos
#llamada a la funcion euler
x, y = EulerMod(f, x0, y0, x1, n)
# print('x = ', x)
# print('y = ', y)                   
```

```python
#Grafica
x2 = np.linspace(x0, x1, 50)
y2 = f2(x2)
plt.figure( figsize = (8, 6) )
plt.plot(x, y, '.--', label = 'Euler Mod.', color = 'blue')
plt.plot(x2, y2, label = 'Funcion', color = 'red')
plt.legend()
plt.grid()
plt.show()
```

##### Dada la ecuación 
\begin{equation}
3\left(1+x^2\right)\frac{dy}{dt}=2xy\left(y^3-1\right),
\end{equation}
##### hallar la solución que satisface la condición inicial 
\begin{equation}
\left. y\right|_{x=0}=2.
\end{equation}
##### Solución analítica 
\begin{equation}
y=\frac{1}{\sqrt{-\frac{7}{8}\left(1+x^2\right)+1}}.
\end{equation}

```python
def f(x,y):
    return 2*x*y**4/(3*(1+x**2))-2*x*y/(3*(1+x**2))

def f2(x):
    return 1/(-7/8*(1+x**2)+1)**(1/3)
```

```python
x0 = 0.0   #valor inicial de tiempo
y0 = 2.0   #valor inicial de la concentracion
x1 = 0.35  #valor final del tiempo
n = 10     #numero de pasos
#llamada a la funcion euler
x, y = EulerMod(f, x0, y0, x1, n)
# print('x = ', x)
# print('y = ', y)
```

```python
#Grafica
x2 = np.linspace(x0, x1, 1000)
y2 = f2(x2)
plt.figure( figsize = (8, 6) )
plt.plot(x, y, '.--', label = 'Euler Mod.', color = 'blue')
plt.plot(x2, y2, label = 'Funcion', color = 'red')
plt.legend()
plt.grid()
plt.show()
```
